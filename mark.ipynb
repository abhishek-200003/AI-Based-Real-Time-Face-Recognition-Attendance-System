```python
import cv2
import os
import numpy as np
import face_recognition as fr
```

```python
path="attendance images"
```

```python
os.listdir(path)
```

```python
pic_list=os.listdir(path)
pic_list
```

```python
input_list=[]
output_list=[]
```

```python
for i in pic_list:
    img=os.path.join(path,i)
    img1=cv2.imread(img)
    input_list.append(img1)
    output_list.append(i.split(".")[0])   
```

```python
print(input_list)
```

```python
print(output_list)
```

```python
def encoder(input_list):
    encoded_img=[]
    for i in input_list:
        conv_img=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_loc= fr.face_locations(conv_img)
        face_en=fr.face_encodings(conv_img,face_loc)[0]
        encoded_img.append(face_en)
    return encoded_img   
```

```python
face_encoding_know_face=encoder(input_list)
```

```python
video = cv2.VideoCapture(0)

while True:

    if 'marked_names' not in globals():
        marked_names = []
    success, img = video.read()
    img1_conv = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    face_loc1 = fr.face_locations(img1_conv)
    face_enco = fr.face_encodings(img1_conv, face_loc1)
    for enc_loc, face_locr in zip(face_enco, face_loc1):
        matches = fr.compare_faces(face_encoding_know_face, enc_loc)
        dist = fr.face_distance(face_encoding_know_face, enc_loc)
        match_index = np.argmin(dist)
        if matches[match_index] == True:
            name = output_list[match_index]
            if name not in marked_names:
                from datetime import datetime
                import csv
                marked_names.append(name)
                now = datetime.now()
                time_string = now.strftime("%H:%M:%S")
                print(name, time_string)
                with open('attendance.csv', 'a', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([name, time_string])
            y1, x2, y2, x1 = face_locr
            cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 3)
            cv2.putText(img, name, (x1,y1),cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 3)
        else:
            name1 = "UNKNOWN"
            print(name1)
            y1, x2, y2, x1 = face_locr
            cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 3)
            cv2.putText(img, name1, (x1,y1),cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 3)
    cv2.imshow("frame", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
